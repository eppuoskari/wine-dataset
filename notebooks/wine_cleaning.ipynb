```python
#Create a column ID for each wine
df['ID'] = range(1, len(df) + 1)

#Style 
df['Style'] = df['Style'].astype(str).str.strip().str.split(' & ')

df = df.explode('Style').reset_index(drop=True)

df['Style'] = df['Style'].str.strip()

print(df.head(20))
```

```python
#Vintage 
df["Vintage"].unique()
```

```python
df = df.rename(columns={"Vintage": "Harvest year"})
df
```

```python
import pandas as pd
import numpy as np

# --- On part de ton df existant ---
# Séparer les années sur "/"
df["Harvest year"] = df["Harvest year"].astype(str).str.split("/")

# Dupliquer uniquement la colonne Harvest year tout en gardant les autres colonnes
df = df.explode("Harvest year", ignore_index=True)

# Nettoyage de la colonne Harvest year
df["Harvest year"] = (
    df["Harvest year"]
    .str.strip()
    .replace({"nan": np.nan, "NV": np.nan})
)

# Conversion au format AAAA
def format_year(val):
    if pd.isna(val):
        return np.nan
    if val.isdigit():
        val_int = int(val)
        if val_int < 100:  # années courtes
            if val_int <= 25:  # ex: "7" ou "21"
                return 2000 + val_int
            else:
                return 1900 + val_int
        return val_int
    return np.nan

df["Harvest year"] = df["Harvest year"].apply(format_year).astype("Int64")

# Aperçu du DF complet avec toutes les colonnes
print(df.head(20))
```

```python
# Characteristics

df["Characteristics"] = df["Characteristics"].fillna("").astype(str)

df["Characteristics"] = df["Characteristics"].str.replace(r"[\[\]']", "", regex=True)

df["Characteristics"] = df["Characteristics"].str.split(",")

df["Characteristics"] = df["Characteristics"].apply(lambda x: [i.strip() for i in x if i.strip() != ""])

df = df.explode("Characteristics", ignore_index=True)

print(df)
```
