```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("/Users/lucielopez/Desktop/stuff/IRONHACK/PROJECT/W4/wine_cleaned_1.csv")
df
```

```python
# Remplacer NaN par chaîne vide et forcer en string
df["Characteristics"] = df["Characteristics"].fillna("").astype(str)

# Retirer crochets, quotes simples/doubles
df["Characteristics"] = df["Characteristics"].str.replace(r"[\[\]']", "", regex=True)

# Séparer sur virgule
df["Characteristics"] = df["Characteristics"].str.split(",")

# Nettoyer espaces et supprimer valeurs vides
df["Characteristics"] = df["Characteristics"].apply(lambda x: [i.strip() for i in x if i.strip() != ""])

# Exploser
df = df.explode("Characteristics", ignore_index=True)

print(df)
```

```python

# Exporter vers un CSV
df.to_csv("wine_clean_2.csv", index=False, encoding="utf-8")


```
